```python
from matplotlib import pyplot as plt 
import numpy as np 
from scipy.optimize import curve_fit
```

```python
# range 30-55
m = [75,78,79,80,81,82,85]
pt_half = [40.17, 41.77, 42.20, 42.77, 43.44, 43.78, 45.47]
pt_half_err = [0.05, 0.05 ,0.05 ,0.05 ,0.05, 0.05, 0.06]


def f(x,a,b):
    return a * x + b

popt, pcov = curve_fit(f,pt_half,m)
t = np.linspace(40,47,1000)
plt.plot(t, f(t,*popt), 'r', label='fit')

plt.errorbar(pt_half, m, xerr=pt_half_err,  fmt='o', markersize=2, color='k', capsize=2, label='data')

print(popt)
print(pcov)

m_real = popt[0] * 43.35 + popt[1]
print('M_w_30-55 =',m_real)
plt.xlabel(r'$p_T$ at half maximum [GeV]')
plt.ylabel(r'$M_W$ [GeV]')
plt.grid()
plt.legend()

m_z = popt[0] * 48.97 + popt[1]
print('m_z_30-55=', m_z)

plt.savefig('gauge1.pdf')
```
